```python
import pandas as pd
import numpy as np
```

```python
df_cities = pd.read_csv('../data/brut/communes-france-2025.csv', index_col=0)

df_cities['dep_code'] = df_cities['dep_code'].astype(str)

df_cities_filtered = df_cities.loc[
    ((df_cities['dep_code'] >= '01') & (df_cities['dep_code'] <= '19'))
    | ((df_cities['dep_code'] >= '21') & (df_cities['dep_code'] <= '95'))
    , [
        'code_insee',
        'nom_standard',
        'dep_code',
        'population',
        'superficie_km2',
        'densite',
        'latitude_mairie',
        'longitude_mairie',
    ]
].sort_values(['population'], ascending=False).reset_index(drop=True)

NB_CHUNKS = df_cities_filtered.shape[0] / 1000

print(NB_CHUNKS)

df_cities_filtered
```
